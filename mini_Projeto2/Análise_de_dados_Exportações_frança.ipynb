### Puxar base de dados

```python
import pandas as pd

tabela = pd.read_csv('exportacoes_franca.csv')

# filtrando as linhas repetidos pegando apenas da Econimic Block da Europa
# tabela = tabela[tabela["Economic Block"] == 'Europe']
tabela = tabela.loc[tabela["Economic Block"] == 'Europe', :]

display(tabela)
```

### Informações Gerais

```python
tabela.info()

# tabela.describe()
```

### Como foi a evolução das exportações para a frança (US$) ao longo dos anos?

```python
exportacao_ano = tabela.groupby("Year").sum()
exportacao_ano = exportacao_ano[['US$ FOB']]

# criando função de formatação
def formatar(valor):
    return f"${valor:,.2f}"

# grafico
import plotly.express as px
exportacao_ano = exportacao_ano.reset_index()
grafico = px.line(exportacao_ano, x="Year", y='US$ FOB', title="Exportações para França por Ano")
grafico.update_xaxes(type="category")
grafico.show()


# aplicando a formatação
exportacao_ano["US$ FOB"] = exportacao_ano["US$ FOB"].map(formatar)

display(exportacao_ano)
```

### Quais os produtos mais exportados ao longo de todo o período?

```python
# Filtrando o que queremos
exportacao_produto = tabela.groupby("SH4 Description").sum()
exportacao_produto = exportacao_produto[["US$ FOB"]]

# Colocando em Ordem decrescente pelo valor US$
exportacao_produto = exportacao_produto.sort_values("US$ FOB", ascending=False)

# Formatando as moeda
exportacao_produto["US$ FOB"] = exportacao_produto["US$ FOB"].map(formatar) 
display(exportacao_produto)
```

### Em 2020 qual cidade mais exportou para a França?

```python
# Filtrando o que queremos
tabela_2020 = tabela.loc[tabela["Year"]==2020, :]

exportacao_cidade_2020 = tabela_2020.groupby("City").sum()
exportacao_cidade_2020 = exportacao_cidade_2020[["US$ FOB"]]

# Colocando em Ordem decrescente pelo valor US$
exportacao_cidade_2020 = exportacao_cidade_2020.sort_values("US$ FOB", ascending=False)

# Formatando as moedas
exportacao_cidade_2020["US$ FOB"] = exportacao_cidade_2020["US$ FOB"].map(formatar) 

display(exportacao_cidade_2020)
```

### O que as 2 maiores cidades exportaram tanto?

```python
# Duque de Caxias (filtrando)
tabela_cidade = tabela_2020.loc[tabela_2020["City"]=="Duque de Caxias - RJ", :]
tabela_cidade = tabela_cidade.groupby("SH4 Description").sum()
tabela_cidade = tabela_cidade[["US$ FOB"]]

# Colocando em Ordem decrescente pelos itens mais vendidos
tabela_cidade = tabela_cidade.sort_values("US$ FOB", ascending=False)

# Formatando as moedas
tabela_cidade["US$ FOB"] = tabela_cidade["US$ FOB"].map(formatar)

display(tabela_cidade[:8]) # Pegando apenas os 8 primeiros
```
